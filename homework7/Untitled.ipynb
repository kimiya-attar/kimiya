```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def make_sine_wave(x, A, w):
    return A * np.sin(w * x)
```

```python
x = np.linspace(0, 2*np.pi, 1000)
```

```python
amplitude = [0.5, 1, 1.5, 2, 2.5]
frequency = [1, 2, 3, 4, 5]
```

```python
fig, axs = plt.subplots(1, 3, figsize=(25, 5))

for A, w in zip(amplitude, frequency):
    # First subplot
    y1 = make_sine_wave(x, A, 1)
    axs[0].plot(x, y1, label=f"A={A}")

    # Second subplot
    y2 = make_sine_wave(x, 1, w)
    axs[1].plot(x, y2, label=f"w={w}")

    # Third subplot
    y3 = make_sine_wave(x, A, w)
    axs[2].plot(x, y3, label=f"A={A}, w={w}")

# Labels
axs[0].set_title("Varying Amplitude")
axs[0].set_xlabel("x")
axs[0].set_ylabel("y")
axs[0].legend()

axs[1].set_title("Varying Frequency")
axs[1].set_xlabel("x")
axs[1].set_ylabel("y")
axs[1].legend()

axs[2].set_title("Varying Amplitude & Frequency")
axs[2].set_xlabel("x")
axs[2].set_ylabel("y")
axs[2].legend()

plt.show()
```
