Most values in the DataFrame are stored as strings (str), even in columns that appear numeric. Some columns, such as “Planet”, “Ring System?”, and “Global Magnetic Field?”, also contain string values representing categorical data.

Storing numeric values as strings prevents mathematical operations from being performed on the data. For example, you cannot compute averages, sums, or other calculations directly. It also affects sorting and analysis, as values are treated as text rather than numbers.

```python
df = pd.read_csv("solar_system.csv")

df = df.set_index("Attribute")
df = df.T
df.index.name = "Planet"
df.reset_index(inplace=True)
df.columns.name = None

for col in df.columns:
    if col not in ["Planet", "Ring System?", "Global Magnetic Field?"]:
        df[col] = pd.to_numeric(df[col], errors="coerce")

def attach_units(col_name, unit, new_name=None):
    df[col_name] = [val * unit for val in df[col_name]]
    if new_name:
        df.rename(columns={col_name: new_name}, inplace=True)

attach_units("Mass (10^24kg)", 1e24 * u.kg, "Mass (kg)")

print(df["Mass (kg)"])
```

```python
df["Semi-Major Axis (10^6 km)"] = (
    df["Perihelion (10^6 km)"] + df["Aphelion (10^6 km)"]
) / 2
```

```python
idx = df.columns.get_loc("Aphelion (10^6 km)")
df.insert(idx + 1, "Semi-Major Axis (10^6 km)", df.pop("Semi-Major Axis (10^6 km)"))
```

```python
print(df.columns)
print(df["Semi-Major Axis (10^6 km)"])
```

```python
df["Orbital Period (years)"] = [
    (val * u.day).to(u.year) for val in df["Orbital Period (days)"]]
```

```python
print(df["Orbital Period (years)"])
```

```python
df.drop(columns=["Orbital Period (days)"], inplace=True)
```

An Astronomical Unit (AU) is the average distance between the Earth and the Sun.

```python
print(const.au)
```

```python
print(const.au.to(u.km))
```

```python
df["Diameter (AU)"] = [(val * u.km).to(u.au) for val in df["Diameter (km)"]]
```

```python
df["Distance from Sun (AU)"] = [(val * 1e6 * u.km).to(u.au) for val in df["Distance from Sun (10^6 km)"]]
```

```python
df["Perihelion (AU)"] = [(val * 1e6 * u.km).to(u.au) for val in df["Perihelion (10^6 km)"]]
```

```python
df["Aphelion (AU)"] = [(val * 1e6 * u.km).to(u.au) for val in df["Aphelion (10^6 km)"]]
```

```python
df["Semi-Major Axis (AU)"] = [(val * 1e6 * u.km).to(u.au) for val in df["Semi-Major Axis (10^6 km)"]]
```

```python
earth = df[df["Planet"] == "Earth"]

print(earth["Diameter (AU)"])
print(earth["Distance from Sun (AU)"])
print(earth["Semi-Major Axis (AU)"])
```
