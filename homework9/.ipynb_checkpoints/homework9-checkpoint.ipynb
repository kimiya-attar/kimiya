```python
def fit_model(x, y):
    p0 = [1, 1]
    popt, pcov = curve_fit(model, x, y, p0=p0)
    A_fit, k_fit = popt

    residuals = y - model(x, A_fit, k_fit)
    chi2 = np.sum(residuals**2)
    dof = len(x) - len(popt)
    chi2_reduced = chi2 / dof

    return A_fit, k_fit, chi2_reduced
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()

noise_columns = ["y_noise_2", "y_noise_3", "y_noise_4", "y_noise_5"]

for i, col in enumerate(noise_columns):
    y = df[col]
    A_fit, k_fit, chi2_reduced = fit_model(x, y)

    axes[i].scatter(x, y, color="gray", label="Noisy data")
    axes[i].plot(x, model(x, A_fit, k_fit), color="red", label="Fitted model")
    axes[i].plot(x, model(x, 5, 0.8), color="black", linestyle="--", label="True model")

    axes[i].set_title(f"{col}, Reduced Chi-Squared = {chi2_reduced:.2f}")
    axes[i].set_xlabel("x")
    axes[i].set_ylabel("y")
    axes[i].legend()

plt.tight_layout()
plt.show()
```

As the noise level increases, the fitted red line deviates more from the true black dotted line because the data becomes more scattered and harder to model accurately.

The reduced chi-squared generally increases as noise increases because the residuals between the data and the model grow larger, indicating a worse fit.

```python
results = []

noise_columns = ["y_noise_1", "y_noise_2", "y_noise_3", "y_noise_4", "y_noise_5"]

for i, col in enumerate(noise_columns, start=1):
    y = df[col]
    A_fit, k_fit, chi2_reduced = fit_model(x, y)

    results.append({
        "Noise level": i,
        "A fit": A_fit,
        "k fit": k_fit,
        "Reduced chi-squared": chi2_reduced
    })

results_df = pd.DataFrame(results)
results_df
```
